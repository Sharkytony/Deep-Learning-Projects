# <div style="background-color: cyan; text-align: center;"><font face="verdana" color="red">Importing Libraries

```python
import os
from PIL import Image
import glob
import numpy as np
import cv2
import imgaug.augmenters as iaa
```

# <div style="background-color: cyan; text-align: center;"><font face="verdana" color="red">Loading Training Images path

```python
train = 'Train_images/'
train_path = os.path.dirname(os.path.abspath(train))+'/'
train_dir_files = os.listdir(train_path)
```

# <div style="background-color: cyan; text-align: center;"><font face="verdana" color="red">Applying Image augmentation on Every Image in Training Images directory

```python
import matplotlib.pyplot as plt

# assigning all the image classes in a list
base_dir = os.listdir('Train_images/')

# looping through each class
for file in base_dir:
    base2 = 'Train_images'

    #connecting the main folder along with sub folder i.e class
    dir_path = os.path.join(base2, file)

    #assigning all images in a list
    dirs = os.listdir(dir_path)    

    #looping through each image in a class
    for item in dirs:

        full_path = os.path.join(dir_path, item)  
        if (os.path.isfile(full_path)) & ('png' in item or 'PNG' in item or 'JPEG' in item or 'jpg' in item):
            im = Image.open(full_path)
            if im.mode != 'RGB':
                im = im.convert('RGB')
            imResize = im.resize((250, 250), Image.LANCZOS)
            imResize.save(full_path, 'JPEG', quality=90)
    
    
    fps = glob.glob(os.path.join(dir_path,'*'))
    images = []
    for fp in fps:
        print(fp) #printing all image paths for verification 
        img = cv2.imread(fp)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        images.append(img_rgb)
    
    # creating an image augmentation model
    preprocessor = iaa.Sequential([
    iaa.Fliplr(),
    iaa.Sometimes(0.1,
        iaa.Grayscale()),
    iaa.Crop(percent=(0,0.1)),
    iaa.Sometimes(0.5,
        iaa.GaussianBlur(sigma=(0, 0.5))
    ),
    iaa.LinearContrast((0.75, 1.5)),
    iaa.AdditiveGaussianNoise(loc=0, scale=(0.0, 0.05*255), per_channel=0.5),
    iaa.Multiply((0.8, 1.2), per_channel=0.2),
    iaa.Affine(
    scale={'x': (0.8, 1.2), 'y':(0.8, 1.2)},
    translate_percent={'x':(-0.2,0.2),'y':(-0.2,0.2)},
    rotate=(-25, 25),
    shear=(-8,8))], random_order=True)
    
    #assingning times each image must be augmented
    aug_times = 8
    
    #looping augmentation model on each image for number of aug_times
    for times in range(aug_times):
        #augmenting image
        images_aug = preprocessor(images=images)
        
        #going through all images 1st time so 0_nameofimage.extension till 8th time i.e.7_nameofimage.extension
        # all images in the img class folder
        i = 0
        for img in images_aug:
            #saving augmented image
            cv2.imwrite(os.path.join(dir_path, f'{times}'+'auged_'+os.path.basename(fps[i])), 
                        cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            i+=1
```
